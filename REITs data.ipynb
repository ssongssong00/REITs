```python
# 리츠 주가 데이터
path1 = '/Users/parkseongwoo/Desktop/REITs/REITs/data/' 
file_list1 = os.listdir(path1)  
file_list_py1 = [file for file in file_list1 if file.endswith('.xlsx')]
```

```python
리츠명 = []
for i in range(len(file_list_py1)):
    a = file_list_py1[i].split('.')[0]
    리츠명.append(a)
```

```python
index = []
for i in range(len(file_list1)):
    index.append(i)
```

```python
리츠주가_df = 
for i, j in zip(file_list_py1,index):
    data1 = pd.read_excel(path1 + i)
    data1 = data1[['일자','종가']]
    data1.rename(columns={'종가':리츠명[j]}, inplace=True)
    리츠주가_df = pd.concat([리츠주가_df,data1],axis = 1)
```
